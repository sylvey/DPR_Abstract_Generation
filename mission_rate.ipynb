```python
with open("data/full_text_no_abstract_ids.txt", "r", encoding="utf-8") as f:
    all_ids = [line.strip() for line in f]

import pandas as pd
df_witout_abstracts = pd.DataFrame(all_ids, columns=["ids"])
```

```python
import pandas as pd

df2 = pd.read_csv('data/binary_data.csv', engine='pyarrow')
df2
```

```python
df_witout_abstracts['ids'] = df_witout_abstracts['ids'].astype(str)
df2['ids'] = df2['ids'].astype(str)
```

```python
import pandas as pd

missing_ids = set(df_witout_abstracts['ids'])
df2['is_missing'] = df2['ids'].astype(str).isin(missing_ids)

label_columns = [col for col in df2.columns if col not in ['ids', 'is_missing', 'dp']]

missing_rates = {}

for label in label_columns:
    label_group = df2[df2[label] == 1]
    
    total_count = len(label_group)
    if total_count > 0:
        missing_count = label_group['is_missing'].sum()
        missing_rates[label] = missing_count / total_count
    else:
        missing_rates[label] = 0.0

df_missing_report = pd.DataFrame.from_dict(missing_rates, orient='index', columns=['missing_rate'])
df_missing_report = df_missing_report.sort_values(by='missing_rate', ascending=False)

print(df_missing_report)
```

```python
# index=True 是因為妳的 label 名稱目前是在 index
df_missing_report.to_csv('missing_rate_report.csv', index=True)
```
