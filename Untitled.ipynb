```python
import pandas as pd

df1_counts = pd.read_csv("data/label_counts_a.csv", index_col=0) ## with
df2_counts = pd.read_csv("data/label_counts_b.csv", index_col=0) ## without

df1_counts["label % in summarization data "] = df1_counts["count"] / df1_counts["count"].sum() * 100
df2_counts["label % w/out abstracts"] = df2_counts["count"] / df2_counts["count"].sum() * 100

df1_counts = df1_counts.reset_index()  # 把原本的 index 轉成 column
df1_counts = df1_counts.rename(columns={"index": "name"})  # 把新產生的欄位改名為 name

df2_counts = df2_counts.reset_index()  # 把原本的 index 轉成 column
df2_counts = df2_counts.rename(columns={"index": "name"})  # 把新產生的欄位改名為 name

df1_counts = df1_counts.drop(columns=["count"])
df2_counts = df2_counts.drop(columns=["count"])

df1_counts
```

```python
df_rouge = pd.read_csv('rouge_scores.csv')
df_rouge
```

```python
df_merged = pd.merge(df_rouge, df1_counts, left_on='labels', right_on='name', how='left')
df_merged
```

```python
df_merged = pd.merge(df_merged, df2_counts, left_on='labels', right_on='name', how='left')
df_merged
```

```python
df_merged = df_merged.drop(columns=["name_x", 'name_y'])
df_merged
```
