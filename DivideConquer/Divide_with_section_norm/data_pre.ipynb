```python
import pandas as pd

df = pd.read_parquet('../dataset/all_articles5-v2_predicted_section_names_lower.parquet')
df
```

```python
df = df.reset_index(drop=True)
df
```

```python
import ast
df['sections'] = df['sections'].apply(ast.literal_eval)
df['sections']
```

```python
dict_rows = []

for i, row in df.iterrows():
    dict_row = {
        'background': '',
        'objective': '',
        'methods': '',
        'results': '',
        'conclusions': '',
        'none': ''
    }

    for text, sec in zip(row['sections'], row['predicted_section_names'].tolist()):
        # print(f"Section: {sec}, Text: {text[:30]}...")
        dict_row[sec] += text
    
    dict_rows.append(dict_row)

df['sections_combined'] = dict_rows
df
```

```python
df.drop(columns=['sections', 'predicted_section_names'], inplace=True)
df
```

```python
df.to_parquet('../dataset/all_articles5-v2_sections_combined.parquet', index=False)
```
